## Schritt 2 – `berechne_drehzahl` mit Default-Parameter

```python
def berechne_drehzahl(drehzahl_an, uebersetzung=1.0):
    return drehzahl_an * uebersetzung

assert berechne_drehzahl(1500) == 1500
assert berechne_drehzahl(1500, 0.5) == 750.0
print("✅")
```

## Schritt 4 – `mittelwert(*werte)`

```python
def mittelwert(*werte):
    if len(werte) == 0:
        return 0
    return sum(werte) / len(werte)

assert mittelwert(2, 4, 6) == 4
assert mittelwert(10, 20) == 15
assert mittelwert() == 0
print("✅")
```

## Mini-Challenge – `max_wert(*werte)`

```python
def max_wert(*werte):
    if len(werte) == 0:
        return None
    return max(werte)

assert max_wert(3, 7, 2, 9, 4) == 9
assert max_wert(-1, -5, -3) == -1
assert max_wert() is None
print("✅")
```
