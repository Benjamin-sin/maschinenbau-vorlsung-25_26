## Aufgabe 1 ⭐ – Fläche eines Rechtecks
**Lösungsidee:** Zwei Parameter multiplizieren und mit `return` zurückgeben.


```python
def flaeche(breite, hoehe):
    return breite * hoehe

assert flaeche(3, 4) == 12
assert flaeche(2.5, 4) == 10.0
assert flaeche(0, 100) == 0
print("✅ Aufgabe 1")
```

## Aufgabe 2 ⭐ – Drehzahl mit Default-Übersetzung
**Lösungsidee:** Default-Wert `1.0` sorgt dafür, dass ohne Übersetzung die Antriebsdrehzahl zurückkommt.


```python
def abtriebsdrehzahl(n_an, uebersetzung=1.0):
    return n_an * uebersetzung

assert abtriebsdrehzahl(1500) == 1500
assert abtriebsdrehzahl(1500, 0.5) == 750.0
assert abtriebsdrehzahl(3000, 2) == 6000
print("✅ Aufgabe 2")
```

## Aufgabe 3 ⭐ – Begrüßung mit Keyword-Argument
**Lösungsidee:** f-String mit den beiden Parametern zusammensetzen.


```python
def begruessung(name, gruss="Hallo"):
    return f"{gruss}, {name}!"

assert begruessung("Anna") == "Hallo, Anna!"
assert begruessung("Ben", gruss="Servus") == "Servus, Ben!"
assert begruessung("Cem", gruss="Moin") == "Moin, Cem!"
print("✅ Aufgabe 3")
```

## Aufgabe 4 ⭐⭐ – Statistik mit `*args`
**Lösungsidee:** `werte` ist ein Tupel. Mit `sum()` und `len()` lassen sich Summe und Mittelwert bilden.


```python
def statistik(*werte):
    if len(werte) == 0:
        return {"anzahl": 0, "summe": 0, "mittelwert": 0}
    s = sum(werte)
    n = len(werte)
    return {"anzahl": n, "summe": s, "mittelwert": s / n}

assert statistik(2, 4, 6) == {"anzahl": 3, "summe": 12, "mittelwert": 4.0}
assert statistik() == {"anzahl": 0, "summe": 0, "mittelwert": 0}
assert statistik(10, 20) == {"anzahl": 2, "summe": 30, "mittelwert": 15.0}
print("✅ Aufgabe 4")
```

## Aufgabe 5 ⭐⭐ – Bauteil-Template mit `**kwargs`
**Lösungsidee:** Datei lesen, dann für jeden Schlüssel `k` in `felder` den Platzhalter
`{K}` (Großbuchstaben) durch den Wert ersetzen.


```python
from pathlib import Path

def fuelle_template(pfad, **felder):
    text = Path(pfad).read_text(encoding="utf-8")
    for k, v in felder.items():
        text = text.replace("{" + k.upper() + "}", str(v))
    return text


ausgabe = fuelle_template(
    "../cad_template_technisch.txt",
    bezeichnung="Zahnrad",
    nummer="ZR-042",
    material="Stahl",
    abmessungen="Ø 50 mm x 12 mm",
    beschreibung="Geradverzahntes Stirnrad, 42 Zähne.",
    bearbeiter="Studentin A",
    datum="2025-11-15",
    status="Freigegeben",
)
print(ausgabe)
assert "Bezeichnung: Zahnrad" in ausgabe
assert "{BEZEICHNUNG}" not in ausgabe
assert "{STATUS}" not in ausgabe
print("✅ Aufgabe 5")
```
