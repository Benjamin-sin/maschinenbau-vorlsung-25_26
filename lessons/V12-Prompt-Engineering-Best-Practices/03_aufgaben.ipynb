## Aufgabe 5 ⭐⭐ – Produktionsüberwachungs-Assistent

**Kontext:** Die Datei [`produktionsdaten_test.csv`](produktionsdaten_test.csv) enthält
pro Stunde eine Zeile mit `Zeit,Stueckzahl,Ausschuss`. Du willst aus den Kennzahlen einen
Prompt bauen, mit dem ChatGPT einen **Tagesbericht** formulieren soll.

**Aufgabe:** Schreibe `produktions_prompt(pfad)`. Die Funktion soll:
1. die CSV-Datei öffnen (Pfad ist `pfad`, `encoding="utf-8"`),
2. die **Header-Zeile überspringen**,
3. alle Stückzahlen und Ausschuss-Werte **aufsummieren** (jeweils `int`),
4. die **Ausschussquote in Prozent** berechnen: `quote = ausschuss / stueck * 100`,
5. folgenden Prompt zurückgeben (Quote auf **2 Nachkommastellen** gerundet, mit `round(...)`):
   ```
   Du bist Produktionsleiter.
   Kennzahlen:
   - Stueckzahl gesamt: <summe_stueck>
   - Ausschuss gesamt: <summe_ausschuss>
   - Ausschussquote: <quote> %
   Aufgabe: Formuliere einen Tagesbericht in 3 Saetzen auf Deutsch.
   ```

**🗺️ Programmablaufplan (PAP):** [diagramme/aufgabe_5_pap.mmd](diagramme/aufgabe_5_pap.mmd)

````python
with open("diagramme/aufgabe_5_pap.mmd", encoding="utf-8") as f:
    display(Markdown(f"```mermaid\n{f.read()}\n```"))
````

````python
def produktions_prompt(pfad):
    summe_stueck = 0
    summe_ausschuss = 0
    # TODO: Datei oeffnen, Header ueberspringen, Zeilen verarbeiten
    quote = 0.0
    return (
        "Du bist Produktionsleiter.\n"
        "Kennzahlen:\n"
        f"- Stueckzahl gesamt: {summe_stueck}\n"
        f"- Ausschuss gesamt: {summe_ausschuss}\n"
        f"- Ausschussquote: {quote} %\n"
        "Aufgabe: Formuliere einen Tagesbericht in 3 Saetzen auf Deutsch."
    )

print(produktions_prompt("produktionsdaten_test.csv"))
````

````python
# ▶️ Selbstkontrolle Aufgabe 5
try:
    # Erwartete Werte unabhaengig aus der Datei berechnen
    s_stueck = 0
    s_aus = 0
    with open("produktionsdaten_test.csv", encoding="utf-8") as f:
        zeilen = f.readlines()
    for zeile in zeilen[1:]:
        zeile = zeile.strip()
        if not zeile:
            continue
        _, stueck, aus = zeile.split(",")
        s_stueck += int(stueck)
        s_aus += int(aus)
    erw_quote = round(s_aus / s_stueck * 100, 2)

    got = produktions_prompt("produktionsdaten_test.csv")
    assert f"Stueckzahl gesamt: {s_stueck}" in got, f"Stueckzahl falsch. Erwartet {s_stueck}, Prompt:\n{got}"
    assert f"Ausschuss gesamt: {s_aus}" in got, f"Ausschuss falsch. Erwartet {s_aus}."
    assert f"Ausschussquote: {erw_quote} %" in got, f"Quote falsch. Erwartet {erw_quote}."
    assert got.startswith("Du bist Produktionsleiter.")
    assert got.rstrip().endswith("auf Deutsch.")
    print("✅ Aufgabe 5 geloest.")
except AssertionError as e:
    print(f"❌ Noch nicht ganz: {e}")
except (NameError, TypeError, FileNotFoundError) as e:
    print(f"❌ Funktion produktions_prompt noch nicht korrekt definiert: {e}")
````
