## Aufgabe 1 ⭐ – CNC-Parameter-Prompt

```python
def baue_cnc_prompt(material, drehzahl, vorschub):
    return (
        "Du bist Fertigungsingenieur.\n"
        f"Aufgabe: Schlage optimale CNC-Parameter fuer {material} vor.\n"
        f"Angaben: Drehzahl={drehzahl} 1/min, Vorschub={vorschub} mm/U.\n"
        "Format: 3 Bullet-Punkte mit Empfehlung und Begruendung."
    )

print(baue_cnc_prompt("Aluminium", 2400, 0.15))
```

## Aufgabe 2 ⭐ – Material-Datenblatt-Zusammenfasser

```python
def datenblatt_prompt(werkstoff, zielgruppe):
    werkstoff = werkstoff.strip()
    return (
        "Du bist technischer Redakteur.\n"
        f"Fasse das Datenblatt zu {werkstoff} zusammen.\n"
        f"Zielgruppe: {zielgruppe}.\n"
        "Format: Markdown-Tabelle | Eigenschaft | Wert | Einheit | (max. 5 Zeilen)."
    )

print(datenblatt_prompt("  S235JR  ", "Konstruktions-Entwurf"))
```

## Aufgabe 3 ⭐ – Fertigungs-Anforderungs-Validator

```python
def validiere_anforderung(text):
    bereinigt = text.strip()
    if not bereinigt:
        return "FEHLER: Anforderung ist leer."
    return (
        "Pruefe die folgende Fertigungs-Anforderung auf Vollstaendigkeit "
        f"(Stueckzahl, Toleranz, Werkstoff): '{bereinigt}'."
    )

print(validiere_anforderung("   "))
print(validiere_anforderung("  Fertige 50 Stueck Welle DIN 668, Toleranz h7.  "))
```

## Aufgabe 4 ⭐⭐ – Few-Shot Zeichnungs-Beschreibung

```python
def baue_few_shot_prompt(eigenschaften):
    beispiele_text = ""
    for e in eigenschaften:
        beispiele_text += f"- {e}\n"
    return (
        "Du bist technischer Zeichner.\n"
        "Beschreibe die technische Zeichnung auf Basis der Eigenschaften.\n"
        "Eigenschaften:\n"
        f"{beispiele_text}"
        "Format: 3 Saetze, einfache Sprache."
    )

print(baue_few_shot_prompt(["Welle", "Durchmesser 20 mm", "Laenge 150 mm", "Passung h7"]))
```

## Aufgabe 5 ⭐⭐ – Produktionsüberwachungs-Assistent

```python
def produktions_prompt(pfad):
    summe_stueck = 0
    summe_ausschuss = 0
    with open(pfad, encoding="utf-8") as f:
        zeilen = f.readlines()
    for zeile in zeilen[1:]:  # Header ueberspringen
        zeile = zeile.strip()
        if not zeile:
            continue
        _, stueck, aus = zeile.split(",")
        summe_stueck += int(stueck)
        summe_ausschuss += int(aus)
    quote = round(summe_ausschuss / summe_stueck * 100, 2)
    return (
        "Du bist Produktionsleiter.\n"
        "Kennzahlen:\n"
        f"- Stueckzahl gesamt: {summe_stueck}\n"
        f"- Ausschuss gesamt: {summe_ausschuss}\n"
        f"- Ausschussquote: {quote} %\n"
        "Aufgabe: Formuliere einen Tagesbericht in 3 Saetzen auf Deutsch."
    )

print(produktions_prompt("../produktionsdaten_test.csv"))
```
