## 🎯 Lernziele-Check
- Titel und Achsen-Labels korrekt setzen.
- Balkenwerte als Liste vollständig angeben.
- Einen eigenen Plot vollständig beschriften.


```python
import matplotlib.pyplot as plt

# Fill-in #1 – Titel ergänzen
messpunkte = ["MP1", "MP2", "MP3", "MP4"]
temperaturen = [62, 71, 58, 68]

plt.figure()
plt.bar(messpunkte, temperaturen, color="steelblue")
plt.title("Lagertemperatur je Messpunkt")
plt.xlabel("Messpunkt")
plt.ylabel("Temperatur in °C")
titel = plt.gca().get_title()
plt.show()

assert titel == "Lagertemperatur je Messpunkt"
print("✅ Fill-in #1 Lösung: Titel =", repr(titel))
```

## Fill-in #2 – X-Achse beschriften

`plt.xlabel("Baugruppe")` ergänzt die fehlende Beschriftung der X-Achse.


```python
import matplotlib.pyplot as plt

baugruppen = ["Rahmen", "Antrieb", "Steuerung", "Hydraulik"]
fehler_pro_100 = [3, 7, 2, 5]

plt.figure()
plt.bar(baugruppen, fehler_pro_100, color="orange")
plt.title("Fehler pro 100 montierte Baugruppen")
plt.xlabel("Baugruppe")
plt.ylabel("Fehler pro 100 Stück")
plt.show()

print("✅ Fill-in #2 Lösung: xlabel='Baugruppe' gesetzt.")
```

## Fill-in #3 – Balkenhöhen eintragen

Die Liste bekommt die vier Quartalswerte in der vorgegebenen Reihenfolge: `[1200, 1350, 1500, 1420]`. Ihre Summe ist `5470`.


```python
import matplotlib.pyplot as plt

quartale = ["Q1", "Q2", "Q3", "Q4"]
stueckzahlen = [1200, 1350, 1500, 1420]

plt.figure()
plt.bar(quartale, stueckzahlen, color="seagreen")
plt.title("Produktion pro Quartal")
plt.xlabel("Quartal")
plt.ylabel("Stückzahl")
plt.show()

assert sum(stueckzahlen) == 5470
print("✅ Fill-in #3 Lösung: Summe =", sum(stueckzahlen))
```

## Mini-Challenge – eigener Plot

Beispiel: Energieverbrauch von vier Antrieben über eine Schicht. Titel, X- und Y-Achse sind gesetzt (Y mit Einheit).


```python
import matplotlib.pyplot as plt

kategorien = ["Antrieb 1", "Antrieb 2", "Antrieb 3", "Antrieb 4"]
werte = [42, 55, 38, 61]  # kWh pro Schicht

plt.figure()
plt.bar(kategorien, werte, color="steelblue")
plt.title("Energieverbrauch pro Antrieb")
plt.xlabel("Antrieb")
plt.ylabel("Energieverbrauch in kWh")
plt.show()

print("✅ Mini-Challenge Lösung: vollständig beschriftet, Summe =", sum(werte), "kWh")
```

## ✅ Zusammenfassung

Jede Lösung folgte dem gleichen Muster: Daten als Listen bereitstellen, `plt.bar(...)` aufrufen, mit `plt.title`, `plt.xlabel` und `plt.ylabel` beschriften, mit `plt.show()` rendern.
