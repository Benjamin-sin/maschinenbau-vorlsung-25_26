## 1. Schleifensteuerung: `break`, `continue`, Loop-`else`

In V07 haben wir drei Werkzeuge kennengelernt, mit denen wir den Ablauf einer Schleife feingranular steuern können. **`break`** bricht die Schleife sofort ab und verlässt sie, auch wenn die Abbruchbedingung noch nicht erfüllt wäre. **`continue`** überspringt nur den aktuellen Durchlauf und springt direkt in die nächste Iteration. Die **`else`**-Klausel einer Schleife — die viele aus anderen Sprachen gar nicht kennen — wird ausgeführt, wenn die Schleife **ohne** `break` regulär durchgelaufen ist.

Diese drei Bausteine kombiniert ergeben ein sehr elegantes Muster für Suchschleifen: „finde das erste Element, das eine Bedingung erfüllt — und gib Bescheid, falls keines gefunden wurde."

> [!NOTE]
> **Loop-`else`** wird **nicht** ausgeführt, wenn die Schleife per `break` verlassen wurde. Diese Asymmetrie ist der ganze Sinn der Klausel.


```python
# Demo: break, continue, loop-else in einer Suchschleife
messwerte = [1.02, 1.05, 0.98, 1.10, 1.22, 1.03]
grenzwert = 1.20

for i, wert in enumerate(messwerte):
    if wert < 0:
        continue  # negative Werte (Fehlmessungen) ueberspringen
    if wert > grenzwert:
        print(f"Grenzwert ueberschritten bei Index {i}: {wert}")
        break
else:
    # wird NUR ausgefuehrt, wenn kein break gefeuert hat
    print("Alle Werte sind im gruenen Bereich.")

print("Schleife beendet.")
```

## 2. List Comprehensions

Eine **List Comprehension** ist eine kompakte Schreibweise, um aus einem bestehenden Iterable (z. B. einer Liste oder einem `range`) eine neue Liste zu erzeugen. Sie ersetzt in vielen Fällen eine klassische `for`-Schleife mit `append` und ist sowohl kürzer zu schreiben als auch schneller auszuführen.

Die Grundform lautet `[ausdruck for element in iterable]`, optional ergänzt um eine Filterbedingung `[ausdruck for element in iterable if bedingung]`. Links vom `for` steht, was mit jedem Element passieren soll; rechts davon, woher die Elemente kommen.

> [!TIP]
> Wenn du gerade erst mit Comprehensions anfängst, schreibe die äquivalente `for`-Schleife darunter in einen Kommentar. So prüfst du, ob Reihenfolge und Filter noch das tun, was du willst.

> [!WARNING]
> Verschachtelte Comprehensions werden schnell unleserlich. Sobald du mehr als zwei `for`-Klauseln oder eine komplexe Bedingung hast, ist eine normale Schleife meist die bessere Wahl.


Drei Muster, die wir später brauchen werden:

1. **Transformation** — jedes Element wird mit einer Rechnung abgebildet: `[w * 2 for w in werte]`.
2. **Filterung** — nur Elemente behalten, die eine Bedingung erfüllen: `[w for w in werte if w > 1.0]`.
3. **Kombination** — Transformation plus Filter in einem Ausdruck: `[w * 2 for w in werte if w > 1.0]`.


```python
# Demo: List Comprehensions
werte = [0.8, 1.05, 0.98, 1.10, 1.22, 0.75]

verdoppelt = [w * 2 for w in werte]
ueber_eins = [w for w in werte if w > 1.0]
prozent_abweichung = [round((w - 1.0) * 100, 1) for w in werte]

print("verdoppelt:       ", verdoppelt)
print("nur ueber 1.0:    ", ueber_eins)
print("Abweichung in %:  ", prozent_abweichung)
```

## 3. Zufallswerte mit dem `random`-Modul

Für Simulationen, Testdaten und einfache Würfel-Logik bringt Python das Modul **`random`** mit. Wir brauchen davon in V14 nur zwei Funktionen: **`random.choice(sequenz)`** wählt ein zufälliges Element aus einer Sequenz, und **`random.randint(a, b)`** liefert eine ganze Zufallszahl zwischen `a` und `b` — Grenzen inklusive.

Wenn du reproduzierbare Ergebnisse brauchst (z. B. für Tests), kannst du mit `random.seed(42)` den Generator auf einen festen Startzustand setzen. Danach liefern alle Funktionen deterministische Folgen.


```python
import random

random.seed(42)  # Reproduzierbarkeit

maschinen = ["Fraese-01", "Drehmaschine-02", "Bohrwerk-03"]
gewaehlte_maschine = random.choice(maschinen)
zufaellige_drehzahl = random.randint(800, 1500)

print(f"Zufaellig ausgewaehlte Maschine: {gewaehlte_maschine}")
print(f"Simulierte Drehzahl:             {zufaellige_drehzahl} U/min")
```

## 4. Listen, Tupel und ihre Methoden

Eine **Liste** (`list`) ist eine veränderbare, geordnete Sammlung von Elementen. Ein **Tupel** (`tuple`) ist das unveränderbare Gegenstück: Du kannst nach Erzeugung keine Elemente mehr hinzufügen, entfernen oder ersetzen. Für V14 reicht das Verständnis: Listen für Sammlungen, die sich ändern dürfen; Tupel für fixe Datensätze wie Koordinaten oder Rückgabewerte aus mehreren Werten.

Die wichtigsten Listen-Methoden sind **`.append(x)`** (hängt `x` hinten an), **`.pop(i)`** (entfernt und liefert das Element an Index `i`, Default: letztes), **`.sort()`** (sortiert in-place), **`.reverse()`** (kehrt in-place um), **`.index(x)`** (liefert den ersten Index von `x`) und **`.count(x)`** (zählt Vorkommen von `x`).

Für den schnellen Zugriff auf Teilbereiche nutzt man **Slicing**: `liste[1:3]` liefert die Elemente an den Indizes 1 und 2, `liste[::-1]` liefert die komplette Liste in umgekehrter Reihenfolge, `liste[::2]` jedes zweite Element. Slicing erzeugt **immer eine Kopie** — die Originalliste bleibt unverändert.

> [!WARNING]
> `.sort()` und `.reverse()` geben **`None`** zurück und arbeiten auf der Liste selbst. `sorted(liste)` und `reversed(liste)` sind die funktionalen Varianten, die eine neue Sammlung liefern.


```python
# Demo: Listen-Methoden und Slicing
messungen = [1.02, 1.05, 0.98, 1.10, 0.98]

# Liste veraendern
messungen.append(1.07)
print("nach append:   ", messungen)

letzte = messungen.pop()
print("pop liefert:   ", letzte, "| Liste:", messungen)

print("Index von 1.05:", messungen.index(1.05))
print("Anzahl 0.98:   ", messungen.count(0.98))

# Slicing (Original bleibt unveraendert)
print("messungen[1:3]:", messungen[1:3])
print("umgekehrt:     ", messungen[::-1])

# Tupel als fixer Datensatz
sensor = ("Temp-01", 22.5, "C")
name, wert, einheit = sensor
print(f"Sensor {name}: {wert} {einheit}")

# sortiert liefert neue Liste; sort() aendert in-place und gibt None zurueck
sortiert = sorted(messungen)
print("sortiert (neue Liste):", sortiert)
```

## 5. Aggregation: `sum`, `all`, `any`, `zip`

Vier eingebaute Funktionen, die wir in jedem Datenauswertungs-Notebook wiedersehen werden. **`sum(iterable)`** addiert alle Zahlen. **`all(iterable)`** liefert `True`, wenn jedes Element truthy ist — praktisch für „alle Bedingungen erfüllt?". **`any(iterable)`** liefert `True`, sobald ein einziges Element truthy ist — praktisch für „mindestens eine Bedingung erfüllt?".

**`zip(a, b)`** paart zwei (oder mehr) Iterables elementweise zu Tupeln. Sobald das kürzere Iterable zu Ende ist, stoppt `zip`. Mit `zip` zusammen mit einer Comprehension lassen sich zwei Listen sauber vergleichen.


```python
# Demo: sum, all, any, zip
messwerte = [1.02, 1.05, 0.98, 1.10, 1.03]
toleranz_unten = 0.95
toleranz_oben = 1.15

print("Summe:        ", sum(messwerte))
print("Durchschnitt: ", round(sum(messwerte) / len(messwerte), 3))

im_toleranz = [toleranz_unten <= w <= toleranz_oben for w in messwerte]
print("alle OK?      ", all(im_toleranz))
print("mind. eine OK?", any(im_toleranz))

# zip: zwei Listen parallel durchgehen
maschinen = ["M1", "M2", "M3"]
auslastung = [0.82, 0.91, 0.77]
for name, last in zip(maschinen, auslastung):
    print(f"  {name}: {last * 100:.0f} %")
```

## 📋 Cheat-Sheet

| Baustein | Typische Verwendung | Mini-Beispiel |
|---|---|---|
| `break` | Schleife sofort abbrechen | `if wert > max: break` |
| `continue` | aktuellen Durchlauf überspringen | `if wert < 0: continue` |
| Loop-`else` | Aktion, wenn Schleife ohne `break` endete | `for ...: ... else: print("ok")` |
| List Comprehension | kompakte neue Liste | `[w * 2 for w in werte]` |
| `random.choice` | Element zufällig wählen | `random.choice(liste)` |
| `random.randint` | Ganzzahl im Bereich | `random.randint(1, 6)` |
| `.append` / `.pop` | Liste verändern | `liste.append(x)` |
| `.sort` / `sorted` | sortieren (in-place / neu) | `sorted(liste)` |
| Slicing | Teilbereich | `liste[1:3]`, `liste[::-1]` |
| `sum` / `all` / `any` | Aggregation | `all(w > 0 for w in werte)` |
| `zip` | zwei Listen parallel durchlaufen | `for a, b in zip(xs, ys): ...` |
