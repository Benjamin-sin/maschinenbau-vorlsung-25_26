## 🎯 Inhalt
- Lösung Fill-in 1: nur Dateinamen aus dem Ordner herausfiltern
- Lösung Fill-in 2: die größte Datei finden

```python
import os
import shutil

# Temp-Ordner in diesem Loesungs-Ordner anlegen, damit das Notebook self-contained ist.
basis = "temp_messdaten"
os.makedirs(basis, exist_ok=True)

dateien = {
    "messung_01.csv": "Zeit,Drehzahl\n0,1200\n1,1210\n2,1205\n",
    "messung_02.csv": "Zeit,Drehzahl\n0,1190\n1,1202\n",
    "messung_03.csv": "Zeit,Drehzahl\n0,1220\n1,1225\n2,1219\n3,1230\n",
    "kalibrierung.txt": "Offset: 0.02\nSteigung: 1.001\n",
    "logbuch.log": "08:00 gestartet\n08:05 Warnung\n",
}
for name, inhalt in dateien.items():
    with open(os.path.join(basis, name), "w", encoding="utf-8", newline="\n") as f:
        f.write(inhalt)

os.makedirs(os.path.join(basis, "archiv"), exist_ok=True)
print(f"Setup fertig: {len(os.listdir(basis))} Eintraege in '{basis}'.")
```

## Fill-in 1 — nur Dateien, keine Unterordner

Wir iterieren über `os.listdir` und testen jeden Eintrag mit `os.path.isfile`. Wichtig: `isfile` braucht den vollständigen Pfad, den wir mit `os.path.join` bauen.


```python
import os

basis = "temp_messdaten"

nur_dateien = [name for name in os.listdir(basis)
               if os.path.isfile(os.path.join(basis, name))]

print("Dateien:", sorted(nur_dateien))

# Selbstkontrolle
erwartet = ["kalibrierung.txt", "logbuch.log", "messung_01.csv", "messung_02.csv", "messung_03.csv"]
assert sorted(nur_dateien) == erwartet
print("✅ Fill-in 1 korrekt.")
```

## Fill-in 2 — die größte Datei finden

Klassisches „Maximum-Suchen-in-Schleife"-Muster: Wir halten die bislang größte Datei in zwei Variablen fest und aktualisieren sie, wenn wir eine größere finden.


```python
import os

basis = "temp_messdaten"

groesste_datei = ""
groesste_byte = 0

for name in os.listdir(basis):
    voller_pfad = os.path.join(basis, name)
    if os.path.isfile(voller_pfad):
        g = os.path.getsize(voller_pfad)
        if g > groesste_byte:
            groesste_byte = g
            groesste_datei = name

print(f"Groesste Datei: {groesste_datei} ({groesste_byte} Byte)")
assert groesste_datei == "messung_03.csv"
assert groesste_byte == 42
print("✅ Fill-in 2 korrekt.")
```

## Mini-Challenge — Summe aller Dateigrößen

Einfache Akkumulation: für jede Datei (also nur echte Dateien, nicht Unterordner) `getsize` aufrufen und aufsummieren.


```python
import os

basis = "temp_messdaten"

summe_bytes = 0
for name in os.listdir(basis):
    voller_pfad = os.path.join(basis, name)
    if os.path.isfile(voller_pfad):
        summe_bytes += os.path.getsize(voller_pfad)

print("Gesamt:", summe_bytes, "Byte")
assert summe_bytes == 164
print("✅ Mini-Challenge korrekt.")

# Alternative mit sum() + Comprehension:
summe_alt = sum(os.path.getsize(os.path.join(basis, n))
                for n in os.listdir(basis)
                if os.path.isfile(os.path.join(basis, n)))
print("Alternative (Comprehension):", summe_alt, "Byte")
```
