## A1 – Paket-Anzahl zählen

Wir lesen die Datei zeilenweise und ziehen die Kopfzeile ab. `readlines()` liefert jede Zeile inklusive Zeilenumbruch als Element zurück.

```python
with open("../netzwerk_pakete.csv", encoding="utf-8") as f:
    zeilen = f.readlines()

paket_anzahl = len(zeilen) - 1  # minus Kopfzeile
print(f"paket_anzahl = {paket_anzahl}")

try:
    assert paket_anzahl == 100
    print("✅ A1 richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```

## A2 – Port prüfen

Zuerst Zeichenart, dann Bereich – so vermeiden wir einen `ValueError` bei `int("abc")`.

```python
def port_valide(port_str):
    if not port_str.isdigit():
        return False
    port = int(port_str)
    return 1 <= port <= 65535

try:
    assert port_valide("80") is True
    assert port_valide("65535") is True
    assert port_valide("abc") is False
    assert port_valide("70000") is False
    assert port_valide("0") is False
    print("✅ A2 richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```

## A3 – IP-Adresse parsen

Mit einer List Comprehension zerlegen und umwandeln in einem Schritt.

```python
def ip_teile(ip_str):
    return [int(x) for x in ip_str.split(".")]

try:
    assert ip_teile("192.168.1.10") == [192, 168, 1, 10]
    assert ip_teile("10.0.0.1") == [10, 0, 0, 1]
    ergebnis = ip_teile("192.168.10.100")
    assert all(isinstance(x, int) for x in ergebnis)
    print("✅ A3 richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```

## A4 – Kritische Maschinen zählen

Bedingung laut PAP: `Latenz > 15 ms` ODER `Fehlerrate > 0.2 %`.

```python
import json

with open("../maschinenkommunikation.json", encoding="utf-8") as f:
    daten = json.load(f)

kritisch_anzahl = 0
for maschine in daten["maschinen"]:
    k = maschine["kommunikation"]
    if k["durchschnittliche_latenz_ms"] > 15 or k["fehlerrate_prozent"] > 0.2:
        kritisch_anzahl += 1
        print(f"  kritisch: {maschine['maschine_id']}")

print(f"kritisch_anzahl = {kritisch_anzahl}")

try:
    assert kritisch_anzahl == 3
    print("✅ A4 richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```

## A5 – Höchster OPC-UA-Messwert

Wir durchlaufen alle `variable`-Elemente, ignorieren Nicht-Double-Werte und sammeln die Maxima.

```python
import xml.etree.ElementTree as ET

baum = ET.parse("../opc_ua_daten.xml")
wurzel = baum.getroot()

hoechster_wert = float("-inf")
for var in wurzel.findall(".//variable"):
    if var.find("data_type").text == "Double":
        wert = float(var.find("value").text)
        if wert > hoechster_wert:
            hoechster_wert = wert

print(f"hoechster_wert = {hoechster_wert}")

try:
    assert hoechster_wert == 8500.0
    print("✅ A5 richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```
