## 1. `with open` – Datei kontrolliert lesen

Der `with`-Block stellt sicher, dass die Datei am Ende automatisch geschlossen wird – auch wenn ein Fehler auftritt. Innerhalb des Blocks ist `f` ein iterierbares Datei-Objekt: Jede Iteration liefert eine Zeile als String inklusive Zeilenumbruch.

```python
with open("sensoren_daten.csv", encoding="utf-8") as f:
    zeilen = f.readlines()

print(f"Anzahl Zeilen inkl. Kopf: {len(zeilen)}")
print(f"Kopfzeile: {zeilen[0].strip()}")
print(f"Erste Datenzeile: {zeilen[1].strip()}")
```

## 2. List- und Dict-Comprehension

Comprehensions sind die Kurzform für Schleifen, die eine neue Liste oder ein neues Dictionary erzeugen. Sie sind kompakt, aber nur dann lesbar, wenn die Logik einfach bleibt.

```python
werte = [1, 2, 3, 4, 5]
quadrate = [w * w for w in werte]
print("Quadrate:", quadrate)

gerade_quadrate = {w: w * w for w in werte if w % 2 == 0}
print("Nur gerade:", gerade_quadrate)
```

## 3. Funktionen mit Default-Parametern

Ein **Default-Parameter** hat im `def` bereits einen Wert zugewiesen. Beim Aufruf kann dieser Wert weggelassen werden – oder per Keyword-Argument gezielt überschrieben werden.

```python
def rabatt(preis, prozent=10):
    """Gibt den Preis nach Abzug des Rabatts zurueck."""
    return preis * (1 - prozent / 100)

print(rabatt(100))           # 10% default
print(rabatt(100, 25))       # 25% per Position
print(rabatt(100, prozent=5))  # 5% per Keyword
```

## 4. Neu in V16: Pandas

**Pandas** ist die Standard-Bibliothek für **Tabellen-Datenanalyse** in Python. Das zentrale Objekt ist das **DataFrame** – eine Tabelle mit benannten Spalten und einem Index. Per Konvention wird `pandas` als `pd` importiert.

> [!NOTE]
> Ein **DataFrame** ist eine zweidimensionale Tabelle mit Spalten unterschiedlicher Datentypen. Eine einzelne Spalte ist eine **Series**. Du kannst dir ein DataFrame wie ein Excel-Blatt vorstellen – nur programmatisch ansteuerbar.

```python
import pandas as pd

df = pd.read_csv("sensoren_daten.csv")
print("Form (Zeilen, Spalten):", df.shape)
df.head(3)
```

Mit `df["Spaltenname"]` greifst du auf eine einzelne Spalte (Series) zu. Darauf stehen dir Aggregations-Methoden wie `.mean()`, `.sum()`, `.min()`, `.max()` direkt zur Verfügung.

```python
print("Mittlere Temperatur:", df["Temperatur_C"].mean())
print("Max. Vibration:", df["Vibration_mm_s"].max())
print("Anzahl Messungen:", df.shape[0])
```
