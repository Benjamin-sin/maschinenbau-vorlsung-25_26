```python
import pandas as pd

df = pd.read_csv("../sensoren_daten.csv")
print("Form:", df.shape)
df.head(2)
```

## Fill-in 1 – Maximum Temperatur

`df["Temperatur_C"].max()` liefert den höchsten Wert direkt.

```python
max_temp = df["Temperatur_C"].max()
print(f"max_temp = {max_temp}")

try:
    assert max_temp == 105.2
    print("✅ Fill-in 1 richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```

## Fill-in 2 – Anzahl heißer Sensoren

Eine Boolean-Maske liefert nur die Zeilen mit `Temperatur_C >= 80`; `len(...)` zählt sie.

```python
anzahl_heiss = len(df[df["Temperatur_C"] >= 80])
print(f"anzahl_heiss = {anzahl_heiss}")

try:
    assert anzahl_heiss == 3
    print("✅ Fill-in 2 richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```

## Fill-in 3 – Typ mit höchstem Mittelwert

`.idxmax()` liefert den Index (also den Gruppennamen) mit dem Maximum der Series.

```python
mittel_pro_typ = df.groupby("Typ")["Temperatur_C"].mean()
bester_typ = mittel_pro_typ.idxmax()
print(f"bester_typ = {bester_typ}")

try:
    assert bester_typ == "Temperatur"
    print("✅ Fill-in 3 richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```

## Mini-Challenge – Top-Vibrations-Typ

Gleicher Ansatz wie oben, nur mit der Spalte `Vibration_mm_s`.

```python
top_vibration_typ = df.groupby("Typ")["Vibration_mm_s"].mean().idxmax()
print(f"top_vibration_typ = {top_vibration_typ}")

try:
    assert top_vibration_typ == "Vibration"
    print("✅ Mini-Challenge richtig.")
except AssertionError as e:
    print(f"❌ {e}")
```
