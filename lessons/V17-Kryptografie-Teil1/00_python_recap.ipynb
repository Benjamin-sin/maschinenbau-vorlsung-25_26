## 1. Strings als Folge von Zeichen

Ein **String** ist eine geordnete Folge einzelner Zeichen. In Python kannst du jedes Zeichen über seinen **Index** ansprechen, wobei die Zählung bei `0` beginnt. Das ist für die Caesar-Chiffre entscheidend, denn dort verschiebst du jeden einzelnen Buchstaben um eine bestimmte Anzahl Positionen im Alphabet.

```python
wort = "MASCHINE"
print("Erstes Zeichen:", wort[0])
print("Drittes Zeichen:", wort[2])
print("Letztes Zeichen:", wort[-1])
print("Länge:", len(wort))
```

### Slicing – Teilabschnitte eines Strings

Mit der Syntax `wort[start:stop]` schneidest du einen Teil-String heraus. Der Start-Index ist **inklusive**, der Stop-Index **exklusive** – das Ergebnis umfasst also `stop - start` Zeichen.

```python
wort = "MASCHINE"
print("wort[0:3] =", wort[0:3])  # 'MAS'
print("wort[3:]  =", wort[3:])   # 'CHINE'
print("wort[:4]  =", wort[:4])   # 'MASC'
```

## 2. Groß- und Kleinschreibung

Die Caesar-Chiffre, die du gleich implementierst, arbeitet nur mit **Großbuchstaben A–Z**. Um einen Eingabetext vorher zu normalisieren, nutzt du `.upper()` (alles groß) oder `.lower()` (alles klein). Beide Methoden verändern den Original-String nicht, sondern geben einen **neuen** String zurück – Strings sind in Python **unveränderlich**.

```python
text = "Hallo Welt"
print(text.upper())  # 'HALLO WELT'
print(text.lower())  # 'hallo welt'
print(text)          # unverändert: 'Hallo Welt'
```

## 3. Iteration über einen String

Eine `for`-Schleife läuft automatisch über jedes einzelne Zeichen eines Strings. Das brauchst du, um Buchstabe für Buchstabe eine Verschlüsselung anzuwenden.

```python
for zeichen in "ABC":
    print("Zeichen:", zeichen)
```

Oft baust du parallel einen **neuen** String auf, indem du einzelne Zeichen mit `+=` anhängst. Dieses Muster siehst du im Caesar-Code dauernd.

```python
quelle = "HALLO"
ergebnis = ""
for zeichen in quelle:
    ergebnis += zeichen + "-"
print(ergebnis)  # 'H-A-L-L-O-'
```

## 4. Der Modulo-Operator `%`

Der Operator `%` liefert den **Rest einer Ganzzahl-Division**. Er ist die mathematische Grundlage der Caesar-Chiffre, denn sobald du über `Z` hinaus verschiebst, musst du vorne wieder bei `A` beginnen – also den Rest modulo 26 bilden.

- `7 % 3` ist `1`, weil `7 = 2 * 3 + 1`.
- `26 % 26` ist `0`, weil 26 glatt durch 26 teilbar ist.
- `28 % 26` ist `2`, weil `28 = 1 * 26 + 2`.
- `3 % 26` ist `3`, weil `3 < 26` ist und der Rest einfach die Zahl selbst bleibt.

```python
print(7 % 3)     # 1
print(26 % 26)   # 0
print(28 % 26)   # 2
print(3 % 26)    # 3
print((25 + 3) % 26)  # 2 – Caesar von Z um 3 Schritte weiter
```

### Integer-Division `//`

Das Gegenstück zum Rest ist die **abrundende Division** `//`, die den ganzzahligen Anteil liefert. In V17 brauchst du `//` nur selten, aber es gehört zum Grundvokabular.

```python
print(7 // 3)    # 2
print(28 // 26)  # 1
print(25 // 26)  # 0
```

## 5. Neu für V17: `ord()` und `chr()`

Jedes Zeichen hat im Computer eine **interne Nummer**, den sogenannten **ASCII-Code**. Die Funktion `ord(zeichen)` liefert diese Nummer, `chr(zahl)` macht den Weg zurück. Mit diesen beiden Funktionen kannst du Buchstaben wie Zahlen behandeln – und das ist exakt das, was die Caesar-Chiffre tut.

Folgende Werte solltest du dir merken:

- `ord("A") == 65`
- `ord("Z") == 90`
- `ord(" ") == 32`
- Die Großbuchstaben `A`–`Z` belegen also den Bereich **65 bis 90**, das sind 26 aufeinanderfolgende Zahlen.

```python
print(ord("A"))   # 65
print(ord("Z"))   # 90
print(chr(65))    # 'A'
print(chr(90))    # 'Z'

# Alle Großbuchstaben per Schleife
for code in range(65, 91):
    print(code, "->", chr(code))
```

### Position im Alphabet (0–25)

Die **Position** eines Großbuchstabens im Alphabet bekommst du, indem du vom ASCII-Code 65 abziehst. `A` hat dann Position 0, `Z` Position 25. Genau mit diesen Positionen rechnet die Caesar-Formel `(pos + k) % 26`.

```python
for zeichen in "ABHZ":
    position = ord(zeichen) - 65
    print(zeichen, "hat Position", position)
```

## ✅ Zusammenfassung

| Werkzeug | Zweck | Beispiel |
|---|---|---|
| `s[i]` | Zeichen an Index `i` | `"HALLO"[0]` → `"H"` |
| `s[a:b]` | Teil-String | `"HALLO"[1:4]` → `"ALL"` |
| `.upper()` / `.lower()` | Groß-/Kleinschreibung | `"abc".upper()` → `"ABC"` |
| `a % b` | Rest der Division | `28 % 26` → `2` |
| `a // b` | Ganzzahlige Division | `28 // 26` → `1` |
| `ord(c)` | Zeichen → ASCII-Code | `ord("A")` → `65` |
| `chr(n)` | ASCII-Code → Zeichen | `chr(65)` → `"A"` |

Mit diesen sieben Werkzeugen baust du gleich eine komplette Caesar-Chiffre zusammen.
