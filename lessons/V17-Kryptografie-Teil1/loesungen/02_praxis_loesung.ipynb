## Hilfsfunktionen aus dem Praxis-Notebook

Damit dieses Notebook eigenständig läuft, definieren wir `caesar_zeichen` und `caesar_text` noch einmal. Inhaltlich 1:1 wie in 02_praxis.ipynb.

```python
def caesar_zeichen(z, k):
    if "A" <= z <= "Z":
        pos = ord(z) - 65
        neu = (pos + k) % 26
        return chr(neu + 65)
    return z

def caesar_text(text, k):
    ergebnis = ""
    for z in text:
        ergebnis += caesar_zeichen(z, k)
    return ergebnis
```

## Fill-in 1 – ASCII-Code vom Buchstaben `M`

Ein direkter `ord()`-Aufruf. Kein Slicing, kein Rechnen, einfach nur die ASCII-Abfrage.

```python
code_m = ord("M")
print("code_m =", code_m)
assert code_m == 77
```

## Fill-in 2 – Buchstabe `T` um 7 verschieben

`T` hat Position 19. `19 + 7 = 26`, und `26 % 26 = 0`. Position 0 ist `A`.

```python
zeichen_7 = caesar_zeichen("T", 7)
print("zeichen_7 =", zeichen_7)
assert zeichen_7 == "A"
```

## Fill-in 3 – `"MASCHINE"` mit Schlüssel 5 verschlüsseln

Ein direkter Aufruf von `caesar_text`. Das Ergebnis ist `"RFXHMNSJ"`.

```python
geheim_maschine = caesar_text("MASCHINE", 5)
print("geheim_maschine =", geheim_maschine)
assert geheim_maschine == "RFXHMNSJ"
```

## Bonus – Lösung der Mini-Challenge

Der im Praxis-Notebook versteckte Geheimtext `"KDOOR YRP JHKHLPHQ GLHQVW"` lässt sich mit Schlüssel `k = 3` entschlüsseln und ergibt `HALLO VOM GEHEIMEN DIENST`. Die Brute-Force-Schleife druckt alle 25 Kandidaten; der lesbare deutsche Satz steht in Zeile `k = 3`.

```python
geheim = "KDOOR YRP JHKHLPHQ GLHQVW"
for k in range(1, 26):
    print(f"k = {k:2d}: {caesar_text(geheim, -k)}")
```

## ✅ Zusammenfassung

- `ord()` liefert den ASCII-Code, `chr()` das Zeichen – beide sind zueinander invers.
- `caesar_zeichen` kapselt die Formel `(pos + k) % 26` für einen einzelnen Buchstaben.
- `caesar_text` wendet das per Schleife auf einen ganzen String an.
