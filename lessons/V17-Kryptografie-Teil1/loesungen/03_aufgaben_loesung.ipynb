## A1 ⭐ – ASCII-Code

Ein reiner Wrapper um `ord()`. Die Funktion übernimmt jedes einzelne Zeichen und liefert den zugehörigen Integer.

```python
def code_von(zeichen):
    return ord(zeichen)

assert code_von("A") == 65
assert code_von("Z") == 90
assert code_von(" ") == 32
assert code_von("a") == 97
print("✅ A1 ok")
```

## A2 ⭐ – Buchstabe verschieben

Drei Schritte: Zeichen → Position (`ord(z) - 65`), Position → neue Position (`% 26`), neue Position → Zeichen (`chr(... + 65)`). Das `% 26` ist entscheidend für den Umlauf am Alphabet-Ende.

```python
def verschiebe(buchstabe, k):
    pos = ord(buchstabe) - 65
    neu = (pos + k) % 26
    return chr(neu + 65)

assert verschiebe("A", 3) == "D"
assert verschiebe("Z", 1) == "A"
assert verschiebe("H", 3) == "K"
assert verschiebe("M", 13) == "Z"
assert verschiebe("A", 0) == "A"
print("✅ A2 ok")
```

## A3 ⭐ – Caesar verschlüsseln

Wir durchlaufen den Text zeichenweise. Ist das Zeichen ein Großbuchstabe (`"A" <= z <= "Z"`), wenden wir `verschiebe` an. Sonst hängen wir es unverändert an. Das `%`-Rechnen steckt komplett in `verschiebe` – `caesar_ver` ist nur die Schleife außen herum.

```python
def caesar_ver(text, k):
    ergebnis = ""
    for z in text:
        if "A" <= z <= "Z":
            ergebnis += verschiebe(z, k)
        else:
            ergebnis += z
    return ergebnis

assert caesar_ver("HALLO WELT", 3) == "KDOOR ZHOW"
assert caesar_ver("ABCXYZ", 1) == "BCDYZA"
assert caesar_ver("HALLO, WELT!", 3) == "KDOOR, ZHOW!"
assert caesar_ver("", 5) == ""
print("✅ A3 ok")
```

## A4 ⭐⭐ – Caesar entschlüsseln

Die kompakteste Lösung nutzt die Symmetrie der Caesar-Chiffre: Um `k` nach rechts verschoben = um `k` nach links zurück. In Python reicht `caesar_ver(text, -k)`, denn der Modulo-Operator behandelt negative Zahlen mathematisch sauber.

Äquivalent wäre `caesar_ver(text, (26 - k) % 26)` – das umgeht Negativ-Modulo und ist lesbar.

```python
def caesar_ent(text, k):
    return caesar_ver(text, -k)

assert caesar_ent("KDOOR ZHOW", 3) == "HALLO WELT"
assert caesar_ent("BCDYZA", 1) == "ABCXYZ"
assert caesar_ent(caesar_ver("MASCHINE AN", 9), 9) == "MASCHINE AN"
assert caesar_ent("KDOOR, ZHOW!", 3) == "HALLO, WELT!"
print("✅ A4 ok")
```

## A5 ⭐⭐ – Brute-Force-Dekodierer

Wir laufen mit `range(1, 26)` über alle 25 Kandidaten-Schlüssel und hängen das jeweilige Entschlüsselungs-Ergebnis an eine Liste. Die **Reihenfolge** ist wichtig: Index 0 entspricht Schlüssel 1, Index 2 entspricht Schlüssel 3 – genau so wie es der Assert fordert.

Eine Kurzform mit **List Comprehension** ist ebenfalls legitim und liefert exakt dieselbe Liste.

```python
def brute_force(geheim):
    ergebnisse = []
    for k in range(1, 26):
        ergebnisse.append(caesar_ent(geheim, k))
    return ergebnisse

# Alternative (List Comprehension):
# def brute_force(geheim):
#     return [caesar_ent(geheim, k) for k in range(1, 26)]

liste = brute_force("KDOOR ZHOW")
assert len(liste) == 25
assert liste[0] == "JCNNQ YGNV"   # Schlüssel 1
assert liste[2] == "HALLO WELT"   # Schlüssel 3
print("✅ A5 ok")
print()
for k, text in enumerate(liste, start=1):
    print(f"k = {k:2d}: {text}")
```

## ✅ Zusammenfassung

- A3 ist die Schleife drüber, A4 die Umkehrung per negativem Schlüssel.
- A5 demonstriert anschaulich, warum Caesar unsicher ist: 25 Versuche genügen, um alles lesbar zu machen.
