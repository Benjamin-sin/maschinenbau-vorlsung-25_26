## Setup – dieselbe In-Memory-DB wie in der Praxis


```python
import sqlite3
import csv

conn = sqlite3.connect(":memory:")
cur = conn.cursor()

cur.execute('''CREATE TABLE Maschinen (
    Maschinen_ID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Typ TEXT NOT NULL,
    Baujahr INTEGER
)''')
cur.execute("INSERT INTO Maschinen VALUES (?, ?, ?, ?)", (1, "Drehmaschine D-7", "Drehen", 2015))
cur.execute("INSERT INTO Maschinen VALUES (?, ?, ?, ?)", (2, "Fraesmaschine F-12", "Fraesen", 2019))
cur.execute("INSERT INTO Maschinen VALUES (?, ?, ?, ?)", (3, "3D-Drucker P-30", "Additiv", 2022))
```

## Lösung 1 – Älteste Maschine


```python
row = cur.execute(
    "SELECT Name FROM Maschinen ORDER BY Baujahr ASC LIMIT 1"
).fetchone()
aeltester_name = row[0]
print(aeltester_name)
assert aeltester_name == "Drehmaschine D-7"
```

## Lösung 2 – Sicheres INSERT


```python
cur.execute(
    "INSERT INTO Maschinen VALUES (?, ?, ?, ?)",
    (4, "Laserschneider L-5", "Laser", 2024),
)
print(cur.execute("SELECT * FROM Maschinen WHERE Maschinen_ID = 4").fetchone())
```

## Lösung 3 – Filter-Query auf Produkte


```python
cur.execute('''CREATE TABLE Produkte (
    Produkt_ID INTEGER PRIMARY KEY,
    Produktname TEXT NOT NULL,
    Produktionszeit_Minuten INTEGER,
    Material_pro_Stueck_kg REAL
)''')
with open("../testdaten/produkte.csv", encoding="utf-8") as f:
    for z in csv.DictReader(f):
        cur.execute(
            "INSERT INTO Produkte VALUES (?, ?, ?, ?)",
            (int(z["Produkt_ID"]), z["Produktname"],
             int(z["Produktionszeit_Minuten"]), float(z["Material_pro_Stueck_kg"])),
        )

schwere_produkte = cur.execute(
    "SELECT Produktname, Material_pro_Stueck_kg FROM Produkte WHERE Material_pro_Stueck_kg > ? ORDER BY Produktname",
    (2.0,),
).fetchall()
for name, masse in schwere_produkte:
    print(f"{name:<20} {masse} kg")
```
